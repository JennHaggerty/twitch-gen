```python
import pandas as pd
import numpy as np
from textgenrnn import textgenrnn

textgen = textgenrnn()
```

```python
data = pd.read_json('test2.txt', orient="columns")
data
```

```python
users = set()
for index, row in data.iterrows():
    users.add(row['commenter']['display_name'])
users
```

```python
EXCLUDE_USERS = ['FeenyBot', 'Nightbot', 'Stay_Hydrated_Bot']
EXCLUDE_LINKS = ['twitch.tv', 'youtube.com']
data = data[[x['display_name'] not in EXCLUDE_USERS for x in data.commenter]]
data = data[[not any([link_st in x['body'] for link_st in EXCLUDE_LINKS]) for x in data.message]]
```

```python
test_list = []
body_list = []
for index, row in data.iterrows():
    body_list.append(row['message']['body'])
    test_list.append(row['message']['fragments'][0]['text'])
print(body_list)
```

```python
np.asarray(body_list)
```

```python
np.savetxt('test-fragment.txt', test_list, fmt='%s', newline = ',', encoding='utf-8')
np.savetxt('test-body.txt', body_list, fmt='%s', newline = ',', encoding='utf-8')
```

```python
textgen.reset()
textgen.train_on_texts(body_list, gen_epochs=2)
```
